```python


# Task: Scrape books under £20 from a website


# Import required libraries
import requests
from bs4 import BeautifulSoup
import csv
-
url = "http://books.toscrape.com/"

response = requests.get(url)

if response.status_code != 200:
    print("Error: Website not accessible")
    exit()
else:
    print("Website accessed successfully")

soup = BeautifulSoup(response.text, "html.parser")

books = soup.find_all("article", class_="product_pod")

# List to store filtered book data
filtered_books = []

for book in books:

    # Extract book title safely
    title = book.h3.a.get("title", "No Title")

    # Find price tag
    price_tag = book.find("p", class_="price_color")

    # Handle missing price tag
    if price_tag is None:
        continue

    try:
        price_text = price_tag.text.replace("£", "")
        price = float(price_text)
    except:
        # Skip if price format is incorrect
        continue

    if price < 20:
        filtered_books.append([title, price])

with open("books_under_20.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Book Title", "Price (£)"])
    writer.writerows(filtered_books)

print("Scraping completed successfully!")
print(f"Total books saved: {len(filtered_books)}")
print("Output file created: books_under_20.csv")
```
